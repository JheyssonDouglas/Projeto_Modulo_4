```python
import pandas as pd
```

```python
df = pd.read_csv('ipsrio.csv')
```

```python
df
```

```python
df.columns
```

```python
central = df[df['zona'] == 'Zona Central']

central
```

```python
maior_ips_central = central[['ips_geral','regiao_administrativa', 'ano']].sort_values(by='ips_geral', ascending=False)

maior_ips_central.groupby('ano').plot.bar(x='regiao_administrativa', y='ips_geral')
```

```python
maior_ips_central.plot.bar(x='regiao_administrativa', y='ips_geral')
```

```python
maior_ips_central.pivot(index='regiao_administrativa', columns='ano', values='ips_geral').plot(kind='bar', figsize= (15, 8))
```
